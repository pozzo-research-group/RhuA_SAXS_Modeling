```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 12
import math
import scipy
import os
```

This notebook is used to model the tube like assembly of the RhuA proteins. The 3d cartesian coordinates of the RhuA protein is first loaded and then it is placed on the coordinates of a 3d helix. The size of the helix can be easily varied based on parameters. Once the tube structure is modeled, its scattering curve can be calculated using the spherical harmonics method. 

First, some functions are defined that will help us in placing each protein in the correct location and orientation.

```python
def helix_coord(amp, f , low, high, n):
    y = np.linspace(low, high, n)
    x = amp*np.cos(f*y)
    z = amp*np.sin(f*y)
    return x, y, z

def relative_coordinates(volume):
    # Determine the points relative to (0,0)
    rel_x = volume[:,0] - np.mean(volume[:,0])
    rel_y = volume[:,1] - np.mean(volume[:,1])
    rel_z = volume[:,2] - np.mean(volume[:,2])
    relative_volume = np.hstack((rel_x.reshape(-1,1), rel_y.reshape(-1,1), rel_z.reshape(-1,1)))
    return relative_volume

def rotation_angle(x,z,r):
    angle = -np.arctan2(z, x)*180/math.pi
    return np.round(angle)

def rotate_coordinates(x, z, angle):
    angle = angle*math.pi/180
    x_new = x*np.cos(angle) + z*np.sin(angle)
    z_new = -x*np.sin(angle) + z*np.cos(angle)
    return x_new, z_new

def place_structure(relative_volume, x,y,z):
    x_coord = relative_volume[:,0] + x
    y_coord = relative_volume[:,1] + y
    z_coord = relative_volume[:,2] + z
    structure_coordinates = np.hstack((x_coord.reshape(-1,1), y_coord.reshape(-1,1), z_coord.reshape(-1,1)))
    return structure_coordinates
```

The 3d cartesian coordinates of the RhuA protein is first loaded and plotted

```python
volume = pd.read_csv('../Data/RhuA_Protein/RhuA_82', delim_whitespace=True).values
volume_rotated = np.hstack((volume[:,0].reshape(-1,1), volume[:,2].reshape(-1,1), volume[:,1].reshape(-1,1)))
```

This plot will show the coordinates of the protein in the "up" position.

```python
# Plot the mean of all points
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.scatter(volume_rotated[:,0],volume_rotated[:,1],volume_rotated[:,2], color = 'blue', alpha = 0.1, s = 30)
ax.set_xlabel('x [$\AA$]')
ax.set_ylabel('y [$\AA$]')
ax.set_zlabel('z [$\AA$]')
#plt.savefig('Figures/RhuA_up.jpg')
```

This plot will show the protein rotated 180 degrees to the "down" position. 

```python
# Plot the mean of all points
volume_xr, volume_zr = rotate_coordinates(volume_rotated[:,0], volume_rotated[:,2], 180)
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.scatter(volume_xr, volume[:,2], volume_zr, color = 'red', alpha = 0.1, s = 30)
ax.set_xlabel('x [$\AA$]')
ax.set_ylabel('y [$\AA$]')
ax.set_zlabel('z [$\AA$]')
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
#plt.savefig('Figures/RhuA_down.jpg')
```

Next, the helical structure will be simulated. This can be done using the mathematical formula for a helix in 3d. 


```python
#Parameters
amp = 215 # Radius of the helix. The mean of the ID and OD of the protein tube 
start = 17 #Number of points in one rotation of the helix

#Fixed Parameters
full_circles = 15 # Number of full circles (15)
low = 0
high = full_circles*80 # Length of Tube, Assuming that one circle is 80 A thick, high = full_circles*80
n = int(start*full_circles+1) # Number of Points 
f = full_circles*2*math.pi/high #frequency

x, y, z = helix_coord(amp, f, low, high, n)
fig = plt.figure(figsize=(8,8))
plt.rcParams['font.size'] = 16
ax = plt.axes(projection='3d')
ax.scatter(x,y,z)
ax.set_xlabel('x', labelpad=10, fontsize=30)
ax.set_ylabel('y', labelpad=10, fontsize=30)
ax.set_zlabel('z', labelpad=10, fontsize=30)
#plt.savefig('../Figures/helix.png', dpi=300, bbox_inches="tight")
```

A RhuA protein is then placed on each coordinate of the helix. The orientation of each protein will be taken into account so that each protein is oriented perpendicular to the y-axis. The orientation angle can be found by the equation:

Once $\theta$ is found, each protein can be orientated using the equation:

The following block of code performs these steps.


```python
# Find the relative volume of the coordinates which is the coordinates centered at 0,0,0
relative_volume = relative_coordinates(volume_rotated)
# Add the relaive coordinates to coordinates on the helix 
angle_lst = []
color_lst = []
for i in range(len(x)):
    if i%2 == 0: #Alternate RhuA structure
        angle = rotation_angle(x[i], z[i], amp) + (90+180)
        color = 1
    else:
        angle = rotation_angle(x[i], z[i], amp) + 90
        color = 2
    x_list = []
    z_list = []
    angle_lst.append(angle)
    for j in range(relative_volume.shape[0]): 
        x_new, z_new = rotate_coordinates(relative_volume[j,0], relative_volume[j,2], angle)
        x_list.append(x_new)
        z_list.append(z_new)
        color_lst.append(color)
    struc_coordinates = np.hstack((np.array(x_list).reshape(-1,1), relative_volume[:,1].reshape(-1,1), np.array(z_list).reshape(-1,1)))
    struc_coordinates = place_structure(struc_coordinates, x[i], y[i], z[i])
    if i == 0:
        coordinates = struc_coordinates
    else:
        coordinates = np.vstack((coordinates, struc_coordinates,))
```

```python
print('Outer Diameter = ', np.max(coordinates[:,0]) - np.min(coordinates[:,0]))
print('Outer Diameter = ', np.max(coordinates[:,2]) - np.min(coordinates[:,2]))
print('Length = ', np.max(coordinates[:,1]) - np.min(coordinates[:,1]))
```

Adjacent proteins will be flipped 180 degrees, since this is the hypothesized conformation.  


```python
protein_1 = coordinates[np.where(np.array(color_lst) == 1)[0], :]
protein_2 = coordinates[np.where(np.array(color_lst) == 2)[0], :]
```

Finally, the tube-like assembly of the RhuA protein is plotted. The proteins are alternating between red and blue for visualization purposes. 

```python
fig = plt.subplots(figsize=(7,7))
#plt.scatter(coordinates[:,1],coordinates[:,2])
plt.scatter(protein_1[:,1], protein_1[:,2], color = 'Red', s = 5, alpha = 0.5)
plt.scatter(protein_2[:,1], protein_2[:,2], color = 'Blue', s = 5, alpha = 0.5)
plt.xlabel('Length of Tube [$\AA$]')
plt.ylabel('Diameter of Tube [$\AA$]')
#plt.xlabel('Length of Tube, y-axis [$\AA$]')
#plt.ylabel('Diameter of Tube, z-axis [$\AA$]')
plt.ylim([-700,700])
plt.xlim([-100,1300])
#plt.xticks([])
#plt.yticks([])
#plt.savefig('Figures/Tube_Side_view_23.png')
plt.show()
```
